```python
import requests
import json
import pandas as pd

# Константы
TOTAL_EPISODS=51

# Класс для общения с API из задания
class APIRequests():
    def __init__ (self):
        self.episode_path='https://rickandmortyapi.com/api/episode/'
        self.character_path='https://rickandmortyapi.com/api/character/'

    def get_episode_json(self, num_episode:str):
        print(requests.get(self.episode_path+num_episode).content)
        return json.loads(requests.get(self.episode_path+num_episode).content)

    def get_origin_name_of_character(self, character_url:str):
        print(requests.get(character_url).content)
        return json.loads(requests.get(character_url).content)['origin']['name']
    
    def get_character_name(self, character_url:str):
        print(requests.get(character_url).content)
        return json.loads(requests.get(character_url).content)['name']

# Класс для формирования витрины данных по заданию
class ETLProcesser():
    def __init__(self):
        self.earth_arr=[]
        self.air_date_arr=[]
        self.req = APIRequests()
    # Основная функция проекта, которая получает необходимую витрину данных
    def earth_by_mounth(self):
        for i in range(1, TOTAL_EPISODS+1):
            episode = req.get_episode_json(str(i))
            #print(episode)
            earth_from_episode = 0
            air_date=episode['air_date']
            for i in range(len(episode['characters'])):
                character_origin_name = req.get_origin_name_of_character(episode['characters'][i])
                if character_origin_name.find('Earth')!=-1:
                    earth_from_episode+=1
            self.earth_arr.append(earth_from_episode)
            self.air_date_arr.append(air_date)
        return [self.air_date_arr, self.earth_arr]
```
